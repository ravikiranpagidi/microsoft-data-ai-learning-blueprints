# Real-Time Events To Lakehouse

Purpose: show the shape of a Fabric notebook that lands generated JSON events into Bronze Delta tables and prepares Silver and Gold outputs. Adjust paths to match your Lakehouse Files area.

```python
from pyspark.sql.functions import col, current_timestamp, input_file_name

source_path = 'Files/logistics/events/*.json'
bronze_table = 'raw_logistics_events'

df = spark.read.option('multiline', 'false').json(source_path)
df = df.withColumn('ingestion_timestamp', current_timestamp()).withColumn('source_file', input_file_name())
df.write.format('delta').mode('append').saveAsTable(bronze_table)
print(f'Loaded {df.count()} rows into {bronze_table}')
```
