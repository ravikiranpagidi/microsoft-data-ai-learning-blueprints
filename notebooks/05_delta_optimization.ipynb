# 05 - Delta Optimization

## Purpose

Show a safe Delta maintenance pattern for Fabric Lakehouse tables. Run maintenance only after understanding retention, recovery, and workload requirements.

```python
gold_tables = ["dim_customer", "dim_account", "dim_product", "dim_branch", "dim_date", "fact_transaction"]

for table_name in gold_tables:
    count_value = spark.table(table_name).count()
    print(f"{table_name}: {count_value} rows")
```

```python
for table_name in gold_tables:
    try:
        print(f"Optimizing {table_name}")
        spark.sql(f"OPTIMIZE {table_name}")
    except Exception as exc:
        print(f"Skipping OPTIMIZE for {table_name}: {exc}")
```

```python
for table_name in gold_tables:
    try:
        print(f"Checking table history for {table_name}")
        display(spark.sql(f"DESCRIBE HISTORY {table_name} LIMIT 5"))
    except Exception as exc:
        print(f"Could not read history for {table_name}: {exc}")
```

## Recommendation

Track job duration, table size, file count, and report query performance before and after changes.
